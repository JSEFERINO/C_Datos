## **a. La Defincición**

+ La distribución de Poisson es una distribución de probabilidad discreta que expresa, a partir de una frecuencia de ocurrencia media, la probabilidad de que ocurra un determinado número de eventos durante cierto período de tiempo.
+ También puede utilizarse para contar el número de eventos o partículas consideradas puntuales en otros tipos de intervalos específicos, como la distancia, el área o el volumen.

+ Fue propuesta por *Siméon-Denis Poisson*, que la dio a conocer en 1838 en su trabajo *Investigación sobre la probabilidad de los juicios en materias criminales y civiles*.

+ Una variable aleatoria discreta $X$ sigue una distribución de Poisson si tiene funcion FMP dada por la formula:
$$ P(X=x)= f(x,λ) = e^{-λ}\frac{λ^x}{x!},  x≥0$$




## **b. Valor Esperado y Varianza de una variable aleatoria discreta de Bernoulli**

+ **Valor Esperado** $$E(X)=\mu=\sum_{x=o}^{∞}x e^{-λ}\frac{λ^x}{x!}$$
+ Se tiene que
$$E(X)=e^{-λ}\sum_{x=o}^{∞}x \frac{λ^x}{x!} \ \
=e^{-λ}\sum_{x=o}^{∞}x \frac{λ^x}{x(x-1)!}
 \ \
=λe^{-λ}\sum_{x=1}^{∞} \frac{λ^{x-1}}{(x-1)!}$$

+ Haciendo el cambio $k=x-1$

$$E(X)=λe^{-λ}\sum_{k=0}^{∞} \frac{λ^k}{k!}=λe^{-λ}e^{λ}=λ$$

+ **Reorganizando**

$$E(X)=\mu=λ$$



## **c. Varianza** $$V(X)=E(X-μ)^2=σ^2=E(X^2)- {\mu}^2$$


$$E(X^2)=\sum_{x=o}^{n}x^2e^{-λ}\frac{λ^x}{x!} \ \ =\sum_{x=o}^{n}[x(x-1)+x]e^{-λ}\frac{λ^x}{x!} $$

$$E(X^2)=\sum_{x=o}^{n}[x(x-1)]e^{-λ}\frac{λ^x}{x!} +\sum_{x=o}^{n}[x]e^{-λ}\frac{λ^x}{x!}$$

$$E(X^2)=\sum_{x=o}^{n}[x(x-1)]e^{-λ}\frac{λ^x}{x(x-1)(x-2)!} +\sum_{x=o}^{n}xe^{-λ}\frac{λ^x}{x!}$$

$$E(X^2)=\sum_{x=2}^{n}λ^2e^{-λ}\frac{λ^{x-2}}{(x-2)!} + E(X)$$

## **d. Función de Distribución Acumulada (FDA)** La función de Distribución acumulada para una variable $$X∼P(λ)$$ viene dada por

Una variable aleatoria discreta $X$ sigue una distribución de Poisson si tiene funcion FMP dada por la formula:
$$ P(X=x)= f(x,λ) = e^{-λ}\frac{λ^x}{x!},  x≥0$$

$$F(x)=P[X\leq x]= \sum_{k=o}^{x}e^{-λ}\frac{λ^k}{k!}$$

## **e. Otras Propiedades**

+ **Función Generadora de Momentos**
 $$M_X(t)=E(e^{tX})=\sum_{x=o}^{∞}e^{tx}e^{-λ}\frac{λ^x}{x!}$$
 $$M_X(t)=E(e^{tX})=e^{-λ}\sum_{x=o}^{∞}\frac{(e^tλ)^x}{x!}$$
 $$M_X(t)=E(e^{tX})=e^{-λ}e^{e^tλ} \ \
 =e^{-λ+e^tλ}=e^{(e^t-1)λ}$$

+ **Función Generadora de Probabilidad** $$G_X(t)=E(t^X)$$

$$G_X(t)=E(t^{X})=\sum_{x=o}^{∞}t^xe^{-λ}\frac{λ^x}{x!}=e^{-λ}\sum_{x=o}^{∞}\frac{(tλ)^x}{x!}=e^{-λ}e^{tλ}\ \
 =e^{-λ+tλ}=e^{(t-1)λ}$$


+ **Función característica** $$\Phi_X(t)=E(e^{itX})=e^{(e^{it}-1)λ}$$

## **f. Propiedades de la función generadora de momentos**

+ **Teorema de Unicidad**.  Si existe $M_X(t)$ de una v.a $X$, determna únivocamente su distribución de probabilidad.
+ **Teorema de relación con los momentos**. Si existe $M_X(t)$ de una v.a $X$ para $t∈(-t_0,t_1)$, se tiene:
+ + Existe el momento $E(X^k)$
+ + $$M_X(t)=\sum_{j=o}^{+∞}\frac{t^j}{j!}E(X^j), t∈(-t_0,t_1)$$
+ + $$E(X^k) =\frac{d^jM_X(t)}{dt^j}|_{t=0}$$

+ **Para variables aleatorias discretas, si existen $$M_X(t)=E(e^{tX}) \ y \ G_X(t)=E(t^{X})$$**, se cumple $$M_X(t)=E(e^{tX})=E((e^t)^{X})=\ G_X(e^t)$$

# **a. Construya el gráfico de la Distribución Poisson dada**

```python
# **a. Graficando la Distribución de Poisson**
plt.figure(figsize=(10, 6))
mu =  10 # parametro de forma
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

# **d. Suponga que la compañia tienen 3 páginas webs y que el número de peticiones promedio de acceso en un minuto cualquiera son $5$, $10$ y $15$, construya en un solo cuadro las distribuciones que representan al número de de peticiones en un minuto cualquiera**

```python
# **d. Comparaciones de Distribución de  Poisson**
plt.figure(figsize=(25, 10))
x=np.linspace(0,50,51)
distributions=pd.DataFrame({'x':x,'$\lambda = 5$':stats.poisson(mu=5).pmf(x),
'$\lambda = 10$':stats.poisson(mu=10).pmf(x),
'$\lambda = 15$':stats.poisson(mu=15).pmf(x),})
ax=distributions.plot(x='x',y='$\lambda = 5$',color='r')
distributions.plot(x='x',y='$\lambda = 10$',color='b',ls='--',ax=ax)
distributions.plot(x='x',y='$\lambda = 15$',color='g',ls='-.',ax=ax)
ax.set_ylim(0, 0.2)
ax.text(8,0.17,'$\lambda=5$')
ax.text(14,0.12,'$\lambda=10$')
ax.text(19,0.10,'$\lambda=15$')
ax.get_legend()
plt.show()
```

```python
# **j. Grafique la Función Generadora de Momentos y la  Función Generadora de Probabilidad**
plt.figure(figsize=(25, 10))
l = 10
x=np.linspace(0,1,1000, dtype = "float")
generadoras=pd.DataFrame({'x':x,'G_Momentos':2.71**((2.71**x-1)*l),'G_Probabilidad':2.71**((x-1)*l)})
ax=generadoras.plot(x='x',y='G_Momentos',color='r')
generadoras.plot(x='x',y='G_Probabilidad',color='b',ls='--',ax=ax)

ax.set_ylim(0, 10)
ax.text(0.3,6,'G_Momentos')
ax.text(0.5,0.5,'G_Probabilidad')

ax.get_legend()
plt.ylabel('observados en y')
plt.xlabel('valores de x')
plt.title('G_Momentos vs G_Probabilidad ')
plt.show()
```
