# **123. Recordemos la Tecnica Bootstrap para el muestreo y estimación de parámetros**
+ BootStrap viene de la expresión "_pulling oneself up by one´s Bootstrap_" que significa "Levantarse a uno mismo jalando la correa de las botas"
+ Para que sirve Bootstrap (Bootstraping) - Para estimar errores estandar e intervalos de confianza independientemente de la distribución de los datos.
+ 

# **a. Técnica de Bootstrapping- Generación Sintética de una Población**

```python
from scipy import stats # ampliamente utilizado
import matplotlib.pyplot as plt # visualizacion
import numpy as np # vectores de datos

np.random.seed(0)

# Detalles de visualización
plt.rcParams["font.size"] = 16
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.left"] = False

# Generación aleatoria de datos sesgados
def datos_sesgados(sesgo, cantidad, maximo):
    aleatorios = stats.skewnorm.rvs(sesgo, 
                                    size=cantidad, 
                                    random_state=0)  # normal sesgada
    aleatorios = aleatorios + abs(aleatorios.min()) 
    aleatorios = np.round(aleatorios, 2)/aleatorios.max() * maximo 
    return np.round(aleatorios, 1)

califs = datos_sesgados(-25.0, 100000, 10) # genere 100000 calificaciones POBLACION

# Visualización de un histograma
plt.hist(califs, bins=11, color="turquoise", edgecolor="black") 
plt.title("Población de Estudiantes\n")
plt.xticks(range(0, 11))
plt.xlabel("\nCalificación")
plt.ylabel("Frecuencia\n")
plt.show()
```

# **b. Extraemos muestras aleatorias de la Población del tamaño 0.1%**

```python
# **b. Extraemos muestras aleatorias de la Población del tamaño 0.1%**
tamano_muestra = 100
muestra_principal = np.random.choice(califs,
                                     tamano_muestra) # choice nos permmite sacar unos datos
muestra_principal
```

# **c. Usando la tecnica de BootStrap seleccionaremos 1000 muestras aleatorias y las guardaremos**

```python
# **c. Usando la tecnica de BootStrap seleccionaremos 1000 muestras aleatorias y las guardaremos**
muestras = np.array([]) #En este espacio guardaremos cada muestra tomada
num_muestras = 1000 # Este será el tamaño de las muestra
for m in range(num_muestras): # Con este procedimiento grantizamos tomar las mil muestras
    muestras = np.append(muestras, 
                         np.random.choice(califs, 
                                          tamano_muestra,
                                          replace=True)) # guardamos en muestras las seleccionadas de la poblacion inicial
muestras = muestras.reshape(-1, tamano_muestra)
muestras.shape, muestras # nos muestras el tamaño del archivo y las muestras
```

# **d. Por cada una de las 1000 muestras de Tamaño 100, se hallan las medias y construye la distribución con estas medias - Distribución Muestral de la Media**

- Como vemos, al no conocer el origen de la muestra, la técnica bootstrap garantiza que la distribución de $\bar{X}$ es normal

```python
# **d. Estamos realizando el muestreo de medias**
plt.hist(muestras.mean(axis=1), bins=100, color="turquoise") # Histograma de la distribución muestral de la media
plt.xlabel("\nCalificaciones promedio de las muestras")
plt.ylabel("Frecuencia\n")
plt.show()
```

# **e. Cálculo del Intervalo de Confianza**

```python
# # **e. Cálculo del Intervalo de Confianza**
intervalo_confianza = np.quantile(muestras.mean(axis=1),
                                  [0.025, 0.975]) # Cuantil 97.5 – Cuantil 2.5 = Intervalo de Confianza del 95%

# Vectores para graficar el intervalo de confianza
x = np.linspace(intervalo_confianza[0], intervalo_confianza[1], 100)
y = [10] * x.size

# Grafica el rango del intervalo de confianza
plt.scatter(x, y, color="deeppink", s=100)

# Grafica el límite inferior del intervalo de confianza
plt.axvline(intervalo_confianza[0], 
            label="IC: Límite inferior",
            color="deeppink",
            lw=4,
            linestyle="--")

# Grafica el límite superior del intervalo de confianza
plt.axvline(intervalo_confianza[1], 
            label="IC: Límite superior", 
            color="deeppink",
            lw=4,
            linestyle="--")

# Grafica la verdadera media poblacional
plt.axvline(califs.mean(), 
            label="Media poblacional",
            color="black",
            lw=10, 
            linestyle=":")

# Grafica la media de la muestra principal
plt.axvline(muestra_principal.mean(), 
            label="Media de muestra principal",
            color="gold",
            lw=10) 

# Grafica la distribución muestral de la media
plt.hist(muestras.mean(axis=1), 
         bins=100, 
         alpha=0.5, 
         color="turquoise") 
print(min(x), max(x))
plt.title("Intervalo de Confianza al 95%\n")
plt.xlabel("\nCalificaciones promedio de las muestras")
plt.ylabel("Frecuencia\n")
plt.legend(bbox_to_anchor=(1, 0.5))
plt.show()
```

# **125. Regresión Lineal Mínimos cuadrados - usando Bootstrap**

# **a. Definamos nuestras variables dependiente e independiente**

```python
# **a. Definamos nuestras variables dependiente e independiente**
import numpy as np
from scipy import optimize
import random

datax = data2['diámetro']
datay = data2['Longitud_ruta']

def f( x, p0, p1):
    return p0 + p1*x

def ff(x, p):
    return f(x, *p)

# Estos son los verdaderos parámetros
p0 = 1.0
p1 = 1.0


# Estas son conjeturas iniciales para ajustes:
pstart = [
    p0 + random.random(),
    p1 + random.random(), 
    
]
```

# **b. El Método es bootstrap**

```python
# **El Método es bootstrap**
def fit_bootstrap(p0, datax, datay, function, yerr_systematic=0.0): # Definimos los parametros de nuestra función 

    errfunc = lambda p, x, y: function(x,p) - y # El error o residuo

    # Ajuste a la primero
    pfit, perr = optimize.leastsq(errfunc, p0, args=(datax, datay), full_output=0) # usamos el metodo de mínimos cuadrados
# ajustamos los datos y residuos

   # Obtener la desviación estándar de los residuales
    residuals = errfunc(pfit, datax, datay) # ingresamos los datos a nuestra funcion creada
    sigma_res = np.std(residuals) # desviación estandar de los residuos

    sigma_err_total = np.sqrt(sigma_res**2 + yerr_systematic**2) # desviación total

    # Aqui usamos bootstrap, 1000 conjuntos de datos aleatorios son generados y ajustados
    ps = [] # aqui almacenaremos esos 1000 conjuntos
    for i in range(1000): # para lista

        randomDelta = np.random.normal(0., sigma_err_total, len(datay)) # Una muestra aleatoria normal 
        randomdataY = datay + randomDelta # Sumamos a nuestra data
        randomfit, randomcov = optimize.leastsq(errfunc, p0, args=(datax, randomdataY), full_output=0)
        ps.append(randomfit) # agregamos al acumulador

    ps = np.array(ps)
    df_data2 =pd.DataFrame(ps)
    mean_pfit = np.mean(ps,0)
    print('Algunas visualizaciones de lo construido hasta ahora:')
    print('Lo almacenado en mean_pfit es ', mean_pfit)
    print('Lo almacenado en ps =',df_data2.describe())
    print('Lo almacenado en randomcov =',randomcov)
    print('Lo almacenado en randomfit =',randomfit)
    

    # Puedes elegir el intervalo de confianza que quieras para tu
     # estimaciones de parámetros: 
    Nsigma = 1. # 1sigma obtiene aproximadamente lo mismo que los métodos anteriores
                 # 1sigma corresponde al intervalo de confianza del 68,3%
                 #2sigma corresponde al intervalo de confianza del 95,44%
    err_pfit = Nsigma * np.std(ps,0) 

    pfit_bootstrap = mean_pfit
    perr_bootstrap = err_pfit
    return pfit_bootstrap, perr_bootstrap 
    print('Lo almacenado en pfit_bootstrap =',pfit_bootstrap)
    print('Lo almacenado en perr_bootstrap =',perr_bootstrap)
    print('Lo almacenado en err_pfit =',err_pfit)
   
pfit, perr = fit_bootstrap(pstart, datax, datay, ff)

print('Lo almacenado en pfit =',pfit)
print('Lo almacenado en perr =',perr)
```

# **c. R2 para LS con BootStrap**

```python
# **c. R2 para LS con BootStrap**
from sklearn.metrics import r2_score
r2 = r2_score(datay, f( datax, *pfit))
print('El coeficiente de Bondad de Ajuste o de Determinación del modelo R_cuadrado = ', r2)
print('Esto es, el modelo explica la variabilidad observada en la respuesta en un porcentaje de ', round(r2*100,1), '%')
```
