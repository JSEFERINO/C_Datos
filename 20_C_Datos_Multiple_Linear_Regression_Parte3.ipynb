* **Forma Matricial del Modelo**

$$Y = X\beta$$
$$ Y = \begin{pmatrix}
y_1\\ y_2 \\ ... \\ y_n \\ \end{pmatrix}; X = \begin{pmatrix}
1 & x_{11} & x_{12} &...& & x_{1k}  \\ 1 & x_{21} & x_{22} &...& & x_{2k} \\ ... \\ 1 & x_{n1} & x_{n2} &...& & x_{nk}\\ \end{pmatrix}; \beta = \begin{pmatrix}
\beta_0\\ \beta_1 \\ ... \\ \beta_k \\ \end{pmatrix}$$

$$\beta = (X^t*X)^{-1}X^tY$$

# **e. Calculando la Variabilidad y el coeficiente de correlacion**

```python
# **e. Calculando la Variabilida y el coeficiente de correlacion**
# Paso 1. Los Datos
x = data2["diámetro"] # variable independiente
y = data2["Longitud_ruta"] # Variable dependiene
n = np.size(x)
  
# Paso 2. Promedios
x_mean = np.mean(x)
y_mean = np.mean(y)
x_mean,y_mean
  
# Paso 3. Variabilidad
Sxy = np.sum(x*y)- n*x_mean*y_mean # Covarianza
Sxx = np.sum(x*x)-n*x_mean*x_mean # Variabilidad en x
Syy = np.sum(y*y)-n*y_mean*y_mean # Variabilidad en x
  
# Paso 4. Coeficientes del modelo
b1 = Sxy/Sxx  # Pendiente
b0 = y_mean-b1*x_mean # intercepto
print('Pendiente b1 es', b1)
print('intercepto b0 es', b0)


# Paso 5. Coeficientes de correlación

r = b1*np.sqrt(Sxx/Syy)  # Coef Correlación de Pearson

print('Coeficientes de correlación r=', r)
print('Coeficientes de Determinación Rcuadrado=', r**2)


# Paso 6. Gráfico de correlación
plt.scatter(x,y)
plt.xlabel('Independent variable X')
plt.ylabel('Dependent variable y')
```


# **f. Dibujando la recta de regresión**

```python
# **f. Dibujando la recta de regresión**
y_pred = b1 * x + b0
  
plt.scatter(x, y, color = 'red')
plt.plot(x, y_pred, color = 'green')
plt.xlabel('X')
plt.ylabel('y')
```

# **g. R cuadrado**

$$SCT = SCM+SCE$$
$$SCT =\sum_{i=1}^n\ {(y_i-\bar{y})^2}$$
$$SCE =\sum_{i=1}^n\ {(y_i-y_{pred})^2}$$
$$R^2 = 1- \frac{SCE}{SCT}$$

```python
# **g. R cuadrado**

E = y - y_pred # Definimos los errores de medicion
SCE = np.sum(E**2) # Variabilidad al cuadrado del Error SCE
print('La variabilidad de error, SCE =', SCE)
  
MSE = SCE/n 
print('Cuadrado Medio del Error MSE =', MSE)
  
RMSE = np.sqrt(mse) # Raíz del cuadrado Medio del Error
print('Raíz del cuadrado Medio del Error RMSE =', RMSE)
  
# Variabilidad Total
SCT = np.sum((y - y_mean)**2) # Variabilidad Total
R2 = 1- (SCE/SCT) # Coeficiente de Determinación
print('Coeficientes de Determinación Rcuadrado=', R2)
```

* **Forma Matricial del Modelo**

$$Y = X\beta$$
$$ Y = \begin{pmatrix}
y_1\\ y_2 \\ ... \\ y_n \\ \end{pmatrix}; X = \begin{pmatrix}
1 & x_{11} & x_{12} &...& & x_{1k}  \\ 1 & x_{21} & x_{22} &...& & x_{2k} \\ ... \\ 1 & x_{n1} & x_{n2} &...& & x_{nk}\\ \end{pmatrix}; \beta = \begin{pmatrix}
\beta_0\\ \beta_1 \\ ... \\ \beta_k \\ \end{pmatrix}$$
$$ \beta = (X^t*X)^{-1}X^tY$$

$$B = \begin{pmatrix}
b_0\\ b_1 \\ ... \\ b_k \\ \end{pmatrix}$$

```python
# **c. Definimos la función para los mínimos cuadrados**
xdata = data2["diámetro"] # Variable independiente
ydata = data2["Longitud_ruta"] # Variable dependiente

def f(x, p0, p1): # recta predicha
    return p1*x + p0 

def ff(x, p):
    return f(x, *p)

# Dos valores para los parámetros
b0 = 5.0
b1 = 5.0

# Estas son conjeturas iniciales para ajustes::
bstart = [
    b0 + random.random(),
    b1 + random.random(), 
    ]

def fit_leastsq(p0, datax, datay, function): # función de minimos cuadrados

    error = lambda p, x, y: y - function(x,p) #diferencia entre el valor observado y el predicho

    pfit, pcov, infodict, errmsg, success = \
        optimize.leastsq(error, p0, args=(datax, datay), \
                          full_output=1, epsfcn=0.0001) # estamos optimizando

    if (len(datay) > len(p0)) and pcov is not None:
        SCE = (error(pfit, datax, datay)**2).sum()/(len(datay)-len(p0))
        pcov = pcov * SCE
        print(SCE)
        print(pcov)
    else:
        pcov = np.inf

    error2 = [] 
    for i in range(len(pfit)):
        try:
          error2.append(np.absolute(pcov[i][i])**0.5)
        except:
          error2.append( 0.00 )
    pfit_leastsq = pfit
    perr_leastsq = np.array(error2) 
    return pfit_leastsq, perr_leastsq 

pfit, perr = fit_leastsq(pstart, xdata, ydata, ff)

print("\n# Fit parameters and parameter errors from lestsq method :")
print("pfit = ", pfit)
print("perr = ", perr)
print(np.cov(xdata,ydata))

```
