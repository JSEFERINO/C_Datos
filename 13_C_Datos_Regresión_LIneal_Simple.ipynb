#**62. Regresión Lineal - Gráfico de Dispersión dos variables**

## **d. Algunos estadisticos de la regresión**

```python
## **d. Algunos estadisticos de la regresión**
pendiente, intercepto, r_value, p_value, std_err = stats.linregress(df_datos2["ESTATURA"],df_datos2["PESO"])
pendiente, intercepto, r_value, p_value, std_err 
```

## **e. Gráfico de Dispersión de las dos variables cuantitativas relacionadas "ESTATURA vs PESO" usando _sns.regplot_ y la recta de regresión**

```python
## **e. Gráfico de Dispersión de las dos variables cuantitativas relacionadas "ESTATURA vs PESO" usando _sns.regplot_ y la recta de regresión**
sns.regplot(x="ESTATURA", y="PESO", scatter=True, order=1, data=df_datos2, label="order 1", line_kws={'label':"y={0:.1f}x+{1:.1f}".format(pendiente, intercepto)})
plt.legend(loc="upper right")
```

## **d. Recta de Regresión PESO vs ESTATURA  para la modalidad FEMENINO**


$$Recta_F(Peso) = a_F + b_F*(Estatura)$$
$$b_F=\frac{cov_F(Peso, Estatura)}{Var_F(Estatura)}$$
$$b_F=\frac{29.322052}{10.429081^2} = 0.27$$
$$a_F =mean_F(Peso)-b_F*mean_F(Estatura) =57.8-0.27*162.4=13.95$$
$$Recta_F(Peso) = 19.95 + 0.27*(Estatura)$$

## **d. Recta de Regresión PESO vs ESTATURA  para la modalidad MASCULINO**


$$Recta_M(Peso) = a_M + b_M*(Estatura)$$
$$b_M=\frac{cov_M(Peso, Estatura)}{Var_M(Estatura)}$$
$$b_M=\frac{47.194623}{11.607576^2} = 0.35$$
$$a_M =mean_M(Peso)-b_M*mean_M(Estatura) 68.949153-0.35*173.762712=8.13$$
$$Recta_M(Peso) = 8.13 + 0.35*(Estatura)$$
