#**58.  Tablas y gráficos de frecuencias usando  _value_counts_ para variables cualitativas**

##**a. Tabla de frecuencias absolutas para una variable _SEXO_**

```python
#**a. Tabla de frecuencias absolutas para una variable _SEXO_**
df_datos2["SEXO"].value_counts()
```

##**b. Grafico de frecuencias absolutas para una variable _SEXO_**:

```python
##**b. Grafico de frecuencias absolutas para una variable _SEXO_**
so.Plot(df_datos2, x="SEXO", color="SEXO").add(so.Bar(), so.Hist())
```

```python
##**b. Tabla de frecuencias relativas para una variable _SEXO_**
df_datos2["SEXO"].value_counts(normalize = True)
```

##**d. Tabla de frecuencias relativas para una variable _SEXO_ redondeadas**

```python
##**d. Tabla de frecuencias relativas para una variable _SEXO_**
df_datos2["SEXO"].value_counts(normalize = True).round(2)
```

##**e. Grafico de frecuencias relativas para una variable _SEXO_**

```python
##**e. Grafico de frecuencias relativas para una variable _SEXO_**
so.Plot(df_datos2, x="SEXO", color="SEXO").add(so.Bar(), so.Hist(stat="proportion"))
```

##**f. Grafico de frecuencias absolutas para dos variables "SEXO vs ESTRATO"**

```python
##**f. Grafico de frecuencias absolutas para dos variables "SEXO vs ESTRATO"**
so.Plot(df_datos2, x="ESTRATO", color="SEXO").add(so.Bar(), so.Hist())
```

##**g. Grafico de frecuencias absolutas para dos variables "SEXO vs ESTRATO"**

```python
##**g. Grafico de frecuencias absolutas para dos variables "SEXO vs ESTRATO"**
so.Plot(df_datos2, x="ESTRATO", color="SEXO").add(so.Bar(), so.Hist(), so.Dodge())
```

##**h. Tabla de frecuencias relativas  para dos variables "SEXO vs ESTRATO"**

```python
##**h. Tabla de frecuencias relativas  para dos variables "SEXO vs ESTRATO"**
df_datos2[["SEXO", "ESTRATO"]].value_counts(normalize = True)
```

##**i. Tabla de frecuencias relativas  para dos variables "SEXO vs ESTRATO" redondeadas**

```python
##**i. Tabla de frecuencias relativas  para dos variables "SEXO vs ESTRATO" redondeadas**
df_datos2[["SEXO", "ESTRATO"]].value_counts(normalize = True).round(3)
```

##**j. Grafico de frecuencias relativas para dos variables "SEXO vs ESTRATO"**

##**k. Tabla de frecuencias absolutas  para tres variables "SEXO vs ESTRATO vs PROGRAMA"**

#**m. Combine con Range para trazar una estimación con barras de error: para dos variables "SEXO vs ESTRATO"**

```python
##**m. Combine con Range para trazar una estimación con barras de error: para dos variables "SEXO vs ESTRATO"**

so.Plot(df_datos2, "DEFINITIVA","SEXO",  color="ESTRATO").add(so.Bar(alpha=.5), so.Agg(), so.Dodge()).add(so.Range(), so.Est(errorbar="sd"), so.Dodge())
```

#**59.  Histogramas de frecuencias usando  para variables cuantitativas**
- Una marca de barra dibujada entre la línea de base y los valores de datos.
- Esta marca dibuja barras entre una línea base y un valor. A diferencia de Bar, las barras tienen un ancho completo y bordes delgados por defecto; esto hace que esta marca sea una mejor opción para un histograma continuo:


## **a. Histograma de frecuencias para la variable cuantitativa Saber 11**

```python
## **a. Histograma de frecuencias para la variable cuantitativa Saber 11**
p = so.Plot(df_datos2, "SABER 11").scale()
p.add(so.Bars(), so.Hist())
```

```python
## **b. Histograma de frecuencias para la variable cuantitativa "Saber 11 vs SEXO**
p.add(so.Bars(), so.Hist(),  so.Stack(),color="SEXO")
```

```python
## **c. Histograma de frecuencias para la variable cuantitativa "Saber 11 vs URBANO**
p.add(so.Bars(), so.Hist(), so.Stack(), color="URBANO")
```

```python
## **d. Histograma de frecuencias para la variable cuantitativa "Saber 11 vs ESTRATO**
p.add(so.Bars(), so.Hist(), so.Stack(), color="ESTRATO")
```
