# **48. Diagrama de cajas para dos variables del Data Frame**

## **a. Importando las librerias a necesitar**

```python
## **a. Importando las librerias a necesitar**
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
```

## **b. BoxPlot para PROMEDIO vs SEXO**

```python
## **b. BoxPlot para PROMEDIO vs SEXO**
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.set_title('PROMEDIO vs SEXO')
sns.boxplot(x="SEXO", y=df_datos2['PROMEDIO'],data=df_datos2,ax=ax)
sns.swarmplot(x="SEXO", y="PROMEDIO", data=df_datos2, color='red', alpha=0.5, ax=ax)
```

## **c. Media y desviación estandard para 'PROMEDIO vs SEXO'**

```python
# **c. Media y desviación estandard para 'PROMEDIO vs SEXO'**
print('PROMEDIO vs SEXO')
df_datos2.groupby('SEXO')['PROMEDIO'].agg(['mean', 'std', 'median', 'count'])
```

## **d. BoxPlot para PROMEDIO vs PROGRAMA**

```python
## **d. BoxPlot para PROMEDIO vs PROGRAMA**
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.set_title('PROMEDIO vs PROGRAMA')
sns.boxplot(x="PROGRAMA", y=df_datos2['PROMEDIO'],data=df_datos2,ax=ax)
sns.swarmplot(x="PROGRAMA", y="PROMEDIO", data=df_datos2, color='red', alpha=0.5, ax=ax)
```

# **e. Media y desviación estandard para 'PROMEDIO vs SEXO'**

```python
# **e. Media y desviación estandard para 'PROMEDIO vs SEXO'**
print('PROMEDIO vs PROGRAMA')
df_datos2.groupby('PROGRAMA')['PROMEDIO'].agg(['mean', 'std', 'median', 'count'])
```

```python
## **f. BoxPlot para PROMEDIO vs CURSO**
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.set_title('PROMEDIO vs CURSO')
sns.boxplot(x="CURSO", y=df_datos2['PROMEDIO'],data=df_datos2,ax=ax)
sns.swarmplot(x="CURSO", y="PROMEDIO", data=df_datos2, color='red', alpha=0.5, ax=ax)
```
